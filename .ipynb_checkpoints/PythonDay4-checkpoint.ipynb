```python
studs =[78,93,145,189,128,102,99]
course=['C','C++','Java','Python','PHP','Javascript','R']
```

```python
plt.title('Total Enrollment')
plt.grid()
plt.xlabel('Course')
plt.ylabel('Enrollement')
plt.bar(course,studs, color='magenta')
```

```python
 stud= pd.read_csv('student.csv')
```

```python
plt.figure(figsize=(16,9))
plt.title('Student performance')
plt.grid()
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.bar(stud['name'],stud['marks'],color='silver')
plt.savefig('students.png')
```

```python
plt.figure(figsize=(16,9))
plt.title('Student performance')
plt.grid()
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.barh(stud['name'],stud['marks'],color='silver')
```

#### Pie Chart

```python
plt.figure(figsize=(10,10))
plt.title('% share of students courses')
plt.pie(studs, labels=course, autopct='%2.2f%%',shadow=True);
```
